```python
import pandas as pd
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("BigDataTask").getOrCreate()
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"

# Download the CSV using pandas and create a Spark DataFrame from it
pd_df = pd.read_csv(url)
df = spark.createDataFrame(pd_df)

df.show()
df.printSchema()
df.count()
df.groupBy("day").sum("total_bill").show()
df.filter(df["total_bill"] > 20).show()
```
